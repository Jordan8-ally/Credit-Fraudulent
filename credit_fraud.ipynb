```python
import pandas as pd
data = pd.read_csv("creditcard.csv")
data.head()
```

```python
data.info()
data.describe()
```

```python
x = data.drop(columns=['Time', 'Class'])
y = data['Class']
print(x.head())
print(y.head())
```

```python
# Plot the features (choose any) and targets (choose any)
import matplotlib.pyplot as plt
plt.scatter(x, y, alpha=0.5, color='red')
plt.xlabel("V1")
plt.ylabel("Amount")
plt.title("Amount vs V1")
plt.show()
```

```python
# Split data into trainning and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
if isinstance(x_train, pd.Series):
    x_train = x_train.to_frame()
if isinstance(x_test, pd.Series):
    x_test = x_test.to_frame()
print("x_train:", x_train.head(10))
print("y_train: ", y_train.head(10))
```

```python
# Scale the feature dataset only after train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Use smote to prevent oversampling
import numpy as np
from imblearn.over_sampling import SMOTE
y_train_class = np.where(y_train > y_train.median(), 1, 0)
smote = SMOTE()
x_resampled, y_resampled = smote.fit_resample(x_train_scaled, y_train_class)
print(x_resampled)
print(y_resampled)
```

```python
# Invoke random forest algorithm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
model = RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42)
model.fit(x_resampled, y_resampled)
y_pred = model.predict(x_test_scaled)
y_prob = model.predict_proba(x_test_scaled)[:, 1]
print("Classification report:\n", classification_report(y_test, y_pred))
print("Confusion matrix: ", confusion_matrix(y_test, y_pred))
print("AUC ROC report: ", roc_auc_score(y_test, y_prob))
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_pred, y_prob)
auc_roc = roc_auc_score(y_test, y_prob)
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='red', lw=2, label=f"ROC Curve (area = {auc_roc:.2f})")
plt.plot([0, 1], [0, 1], color='blue', lw=2, linestyle='--', label="Random Classifier")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.title("Receiver Operating Characteristic Curve")
plt.legend(loc='lower right')
plt.show()
```

```python
import seaborn as sns
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, fmt='d', annot=True, cmap='coolwarm')
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix")
plt.show()
```
